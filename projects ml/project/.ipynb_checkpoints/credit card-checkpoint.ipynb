# Data Engineering

```python
#label encoded data for tree based algos
def label_encode(col):
    le = LabelEncoder()
    le.fit(col)
    print(col.name)
    return le.transform(col)
```

```python
for i in obj_cols:
    bank[i] = label_encode(bank[i])
```

```python
bank.head()
```

```python
y = bank["y"]
x = bank.drop(["y"], axis=1)
```

```python
# outliers quantiles.
q1 = x.quantile(0.75)
q2 = x.quantile(0.25)
iqr = q1 - q2
ul = q1 + 1.5 * iqr
ll = q2 - 1.5 * iqr
print(ul, "\n")
print(ll)
```

```python
((x>ul) | (x<ll)).sum()
```

```python
x[((x>ul) | (x<ll))] = np.NAN
```

```python
#appying KNN amputation
#filling outliers with nan values
from sklearn.impute import KNNImputer
imputer = KNNImputer()
```

```python
#replacing Nan values with their nearest neigbour value
bank_imputed = pd.DataFrame(imputer.fit_transform(x), columns=x.columns)
bank_imputed.head()
```

```python
# Splitting test and traingin data 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(bank_imputed, y, test_size = 0.2, stratify=y, random_state=42)
```

```python
# building enseble model
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
rf.fit(x_train, y_train)
```

```python
rf.fit(x_train, y_train)
```

```python
pd.DataFrame({"features": x_train.columns, "importance": rf.feature_importances_}).sort_values(by="importance", ascending=False)
```

```python
pred_train_rf = rf.predict(x_train)
pred_test_rf = rf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print("..Confusion Matrix Train..")
print(confusion_matrix(y_train, pred_train_rf))
print("..Confusion Matrix Test..")
print(confusion_matrix(y_test, pred_test_rf))
```

```python
print("..accuracy_score Train..")
print(accuracy_score(y_train, pred_train_rf))
print("..Confusion Matrix Test..")
print(accuracy_score(y_test, pred_test_rf))
```

```python
print("..classification_report Train..")
print(classification_report(y_train, pred_train_rf))
print("..Confusion Matrix Test..")
print(classification_report(y_test, pred_test_rf))
```
